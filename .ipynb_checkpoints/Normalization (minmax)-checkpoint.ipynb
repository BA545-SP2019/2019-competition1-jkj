```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
# Uploading imputed data to df from CSV WorkingDataFrame
df = pd.read_csv('WorkingDataFrame', index_col=0)
df.head()
```

```python
# Taking catagorical variables and putting them in df_I
# This way, they do not interfere with Normalization
df_I = df.iloc[:,:3]
df_I.head()
```

```python
# Temporarily dropping catagorical variables from 
# norm_df DataFrame 
df=df.drop(['I1', 'I2', 'I3'], axis=1)
df.head()
```

```python
# Taking target variables and putting them in df_T
# This way, they do not interfere with Normalization
df_T = df.iloc[:,19:21]
df_T.head()
```

```python
# Temporarily dropping catagorical variables from 
# norm_df DataFrame 
df=df.drop(['Y1','Y2'], axis=1)
df.head()
```

```python
drop_df = pd.concat([df_I, df_T], axis=1, join_axes=[df.index])
drop_df.head()
```

```python
# Using Normalized Data, creating 6 new columns 
# These columns will replace the numerical values of T1-T5 and S1-S3 with ratios 

# Number of Long Sentences / Number of Long Sentences = T4' 
norm_df["T4'"]=norm_df['T4']/df['T1']

# Number of Read Words / Number of Words = T3' 
norm_df["T3'"]=norm_df['T3']/df['T2']

# Number of Long Words / Number of Words = T5'
norm_df["T5'"]=norm_df['T5']/df['T2']

# Number of Positive Words / Number of Words = S1'
norm_df["S1'"]=norm_df['S1']/df['T2']

# Number of Negative Words / Number of Words = S2'
norm_df["S2'"]=norm_df['S2']/df['T2']

# Number of Uncertain Words / Number of Words = S3'
norm_df["S3'"]=norm_df['S3']/df['T2']

# Checking that the new attributes are in the norm_df DataFrame
norm_df.head()
```

```python
# Observing the mean of the norm_df DataFrame 
norm_df.mean()

# TRY THIS AGAIN WHEN THE VALUES ARE NORMALIZED 
```

```python
# Adding the three catagorical values back into the DataFrame 
norm_df = pd.concat([drop_df, norm_df], axis=1, join_axes=[df.index])

# Checking the merge 
norm_df.head()
```
