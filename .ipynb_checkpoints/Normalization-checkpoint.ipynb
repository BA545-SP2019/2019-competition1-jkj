```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
df = pd.read_csv('WorkingDataFrame', index_col=0)
df.head()
```

```python
df_I = df.iloc[:,:3]
df_I.head()
```

```python
df=df.drop(['I1', 'I2', 'I3'], axis=1)
df.head()
```

```python
def normalize(dataset): 
    norm_df = (df - df.mean()) / (df.max() - df.min())
    return norm_df
norm_df = normalize(df)
norm_df.head()
```

```python
full_df = pd.concat([df_I, norm_df], axis=1, join_axes=[df.index])
full_df.head()
```

```python
# Using Normalized Data, creating 6 new columns 
# These columns will replace the numerical values of T1-T5 and S1-S3 with ratios 
norm_df["T4'"]=norm_df['T4']/norm_df['T1']
norm_df["T3'"]=norm_df['T3']/norm_df['T2']
norm_df["T5'"]=norm_df['T5']/norm_df['T2']
norm_df["S1'"]=norm_df['S1']/norm_df['T2']
norm_df["S2'"]=norm_df['S2']/norm_df['T2']
norm_df["S3'"]=norm_df['S3']/norm_df['T2']
norm_df.head()
```

```python
norm_df.mean()
```

```python
norm_df = pd.concat([df_I, norm_df], axis=1, join_axes=[norm_df.index])
norm_df.head()
```

```python
# Exporting DataFrame as CSV 
norm_df.to_csv('NormDF', ',')
```
