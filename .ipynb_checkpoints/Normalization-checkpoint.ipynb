```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
df = pd.read_csv('WorkingDataFrame', index_col=0)
df.head()
```

```python
df_I = df.iloc[:,:3]
df_I.head()
```

```python
df=df.drop(['I1', 'I2', 'I3'], axis=1)
df.head()
```

```python
def normalize(dataset): 
    df_norm = (df - df.mean()) / (df.max() - df.min())
    return df_norm
df_norm = normalize(df)
df_norm.head()
```

```python
full_df = pd.concat([df_I, df_norm], axis=1, join_axes=[df.index])
full_df.head()
```

```python
# Exporting DataFrame as CSV 
full_df.to_csv('NormDF', ',')
```
