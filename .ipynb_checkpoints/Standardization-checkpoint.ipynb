```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
norm_df = pd.read_csv('NormDF', index_col=0)
norm_df.head()
```

```python
df_I = norm_df.iloc[:,:3]
df_I.head()
```

```python
norm_df=norm_df.drop(['I1', 'I2', 'I3'], axis=1)
norm_df.head()
```

```python
# Get column names first
norm_names = norm_df.columns
# Create the Scaler object
scaler = preprocessing.StandardScaler()
# Fit your data on the scaler object
stand_df = scaler.fit_transform(norm_df)
stand_df = pd.DataFrame(stand_df, columns=norm_names)
```

```python
print(stand_df)
```

```python
stand_df = pd.concat([df_I, stand_df], axis=1, join_axes=[stand_df.index])
stand_df.head()
```

```python
stand_df.max()
```

```python
stand_df.mean()
```

```python
stand_df.std()
```

```python
# Exporting DataFrame as CSV 
stand_df.to_csv('StandDF', ',')
```
