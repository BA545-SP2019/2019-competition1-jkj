```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
norm_df = pd.read_csv('NormDF', index_col=0)
norm_df.head()
```

```python
df_I = norm_df.iloc[:,:3]
df_I.head()
```

```python
norm_df=norm_df.drop(['I1', 'I2', 'I3'], axis=1)
norm_df.head()
```

```python
# Get column names first
norm_names = norm_df.columns
# Create the Scaler object
scaler = preprocessing.StandardScaler()
# Fit your data on the scaler object
scaled_df = scaler.fit_transform(norm_df)
scaled_df = pd.DataFrame(scaled_df, columns=norm_names)
```

```python
print(scaled_df)
```

```python
scaled_df = pd.concat([df_I, scaled_df], axis=1, join_axes=[scaled_df.index])
scaled_df.head()
```

```python
scaled_df.max()
```

```python
scaled_df.mean()
```

```python
scaled_df.std()
```

```python
# Exporting DataFrame as CSV 
scaled_df.to_csv('StandDF', ',')
```
