```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
%matplotlib inline
plt.style.use('ggplot')
```

```python
# Creating DataFrame df from CSV SkewDF2
df = pd.read_csv('SkewDF2', index_col=0)

# Displaying first five results 
df.head()
```

```python
# Using Imputed Data, creating 6 new columns 
# These columns will replace the numerical values of T1-T5 and S1-S3 with ratios 

# Number of Long Sentences / Number of Long Sentences = T4' 
df["T4'"]=df['T4']/df['T1']

# Number of Read Words / Number of Words = T3' 
df["T3'"]=df['T3']/df['T2']

# Number of Long Words / Number of Words = T5'
df["T5'"]=df['T5']/df['T2']

# Number of Positive Words / Number of Words = S1'
df["S1'"]=df['S1']/df['T2']

# Number of Negative Words / Number of Words = S2'
df["S2'"]=df['S2']/df['T2']

# Number of Uncertain Words / Number of Words = S3'
df["S3'"]=df['S3']/df['T2']

# Checking that the new attributes are in the norm_df DataFrame
df.head()
```

```python
df = df.drop(['T1','T2','T3','T4','T5','S1','S2', 'P(mid)'], axis=1)
```
