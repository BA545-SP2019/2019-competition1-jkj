# Data Exploration 

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
```

```python
# Importing Excel file, calling the DataFrame comp_df
comp_df = pd.read_excel("Competition1_raw_data_fixed.xlsx")

# In Excel, we edited column names to omit paranetheses. 
# We also handled 8 missing values in the I3 columns using SICCODE.com to find missing values 
# For those with multiple SIC codes, we chose the primary one - found from SICCODE.com
# For those with NAICS codes, we converted them to SIC codes. 

# Checking import 
comp_df.head(15)
```

```python
# Checking out the data
comp_df.info()

# This is not taking into account missing values
# 682 is the total amount of rows in the Dataset as confirmed with the Excel file 
# This can be further seen looking at the tenth row 
```

```python
# Having Python recognize that there are missing values in the dataset. 

# Since Python is not recognizing the "-" as missing values, we are replacing said values.
# Using np.nan as the value, the dataset recognizes that the value is "not a number"
comp_df.replace(to_replace="-", value=np.nan, inplace=True)
```

```python
#Checking Row 10 now, the "-" are replaced with NaN 
comp_df.head(15)
```

```python
#By changing the value of the "-" in the DataFrame, Python was able to recognize 
#that there are missing values and the datatypes are NOT non-null objects
#But rather the first three columns are non-null objects and the rest are floats.
comp_df.info()
```

```python
# checking for null values 
print(comp_df.isnull().sum(axis=0).tolist())
```

# Imputation of Columns

```python
#Using MEDIAN of column to fill in Continuous/Float columns
# We are using the median because the columns are not normally distributed
#Using MODE of column to fill in Binary/Boolean columns 
comp_df['P_IPO'].fillna(comp_df['P_IPO'].mean(), inplace=True)
comp_df['P_H'].fillna(comp_df['P_H'].mean(), inplace=True)
comp_df['P_L'].fillna(comp_df['P_L'].mean(), inplace=True)
comp_df['P_1DAY'].fillna(comp_df['P_1DAY'].mean(), inplace=True)
comp_df['C1'].fillna(comp_df['C1'].mean(), inplace=True)
comp_df['C2'].fillna(comp_df['C2'].mode()[0], inplace=True)
comp_df['C3'].fillna(comp_df['C3'].mean(), inplace=True)
comp_df['C4'].fillna(comp_df['C4'].mean(), inplace=True)
comp_df['C5'].fillna(comp_df['C5'].mean(), inplace=True)
comp_df['C6'].fillna(comp_df['C6'].mean(), inplace=True)
comp_df['C7'].fillna(comp_df['C7'].mean(), inplace=True)
comp_df['T1'].fillna(comp_df['T1'].mean(), inplace=True)
comp_df['T2'].fillna(comp_df['T2'].mean(), inplace=True)
comp_df['T3'].fillna(comp_df['T3'].mean(), inplace=True)
comp_df['T4'].fillna(comp_df['T4'].mean(), inplace=True)
comp_df['T5'].fillna(comp_df['T5'].mean(), inplace=True)
comp_df['S1'].fillna(comp_df['S1'].mean(), inplace=True)
comp_df['S2'].fillna(comp_df['S2'].mean(), inplace=True)
comp_df['S3'].fillna(comp_df['S3'].mean(), inplace=True)
comp_df.head(15)
```

```python
# checking for null values 
print(comp_df.isnull().sum(axis=0).tolist())
```

# Creating New 'Ratio' Columns

```python
# Checking for 0 values in T1 to avoid "inf" errors later on with new ratio variables
# If this problem is not resolved up front in the notebook 
comp_df.loc[comp_df['T1'] < 0]
# No values in T1 are 0, therefore none have to be imputed
```

```python
# Checking for 0 values in T2 to avoid "inf" errors later on with new ratio variables
comp_df.loc[comp_df['T2'] < 0]

# Replacing the 0 values with the median of that column 
comp_df['T2'] = comp_df.loc[(comp_df['T2']<0)] = comp_df['T2'].median()
```

```python
# Checking for 0 values in T2 to avoid "inf" errors later on with new ratio variables
comp_df.loc[comp_df['T2'] < 0]
```

```python
# Checking for -1 values in T1 to avoid "neg" errors later on with new ratio variables
comp_df.loc[comp_df['T5'] < 0]
# Changed negative value in Row 7 to Median of T5 on Excel 
comp_df.T5.mean()
```

```python
# Checking for -1 values in T1 to avoid "neg" errors later on with new ratio variables
comp_df.loc[comp_df['S1'] < 0]
# Changed negative value in Row 134 to Median of S1 on Excel 
comp_df.S1.mean()
```

```python
# Creating Target Variable Y1 = Pre-IPO Price Revision
comp_df['Y1']= np.where(comp_df['P_IPO']<((comp_df['P_H']+comp_df['P_L'])/2), 1, 0)
comp_df.head()
```

```python
# Creating Target Variable Y2 = Post-IPO Initial Return 
comp_df['Y2']= np.where(comp_df['P_IPO']<comp_df['P_1DAY'], 1, 0)
comp_df.head()
```

```python
# Creating Control Variable C3' = Positive EPS Dummy
comp_df["C3'"]=np.where(comp_df['C3']>0, 1, 0)
comp_df.head()
```

```python
# Creating Control Variable C5' = Share Overhang
comp_df["C5'"]=(comp_df['C5']/comp_df['C6'])
comp_df.head(2)
```

```python
# Creating P(mid) Variable to help with the creation of the Control Variable C6'
comp_df["P(mid)"]=(comp_df['P_H']+comp_df['P_L'])/2
comp_df.head(2)
```

```python
# Creating Control Variable C6' = Up Revision 
comp_df["C6'"]=np.where(comp_df['P_IPO']>comp_df["P(mid)"], (comp_df["P_IPO"]-comp_df["P(mid)"])/comp_df["P(mid)"]*100,0)
comp_df.head(2)
```

```python
# Now that the prime values for C3, C5 and C6 have been made we can drop the originals
comp_df = comp_df.drop(["C3","C5","C6","P(mid)"], axis=1)
comp_df.head(2)
```

```python
# Using Imputed Data, creating 6 new columns 
# These columns will replace the numerical values of T1-T5 and S1-S3 with ratios 

# Number of Long Sentences / Number of Long Sentences = T4' 
comp_df["T4'"]=comp_df['T4']/comp_df['T1']

# Number of Read Words / Number of Words = T3' 
comp_df["T3'"]=comp_df['T3']/comp_df['T2']

# Number of Long Words / Number of Words = T5'
comp_df["T5'"]=comp_df['T5']/comp_df['T2']

# Number of Positive Words / Number of Words = S1'
comp_df["S1'"]=comp_df['S1']/comp_df['T2']

# Number of Negative Words / Number of Words = S2'
comp_df["S2'"]=comp_df['S2']/comp_df['T2']

# Number of Uncertain Words / Number of Words = S3'
comp_df["S3'"]=comp_df['S3']/comp_df['T2']

# Checking that the new attributes are in the norm_df DataFrame
comp_df.head()
```

```python
# Now that the prime values for the T's and S's have been made we can drop the originals
comp_df = comp_df.drop(["T1","T2",'T3','T4','T5','S1','S2','S3'], axis=1)
comp_df.head(2)
```

```python
comp_df.info()
# Checking remaining columns
# Also checking to make sure there are no null values left 
```
