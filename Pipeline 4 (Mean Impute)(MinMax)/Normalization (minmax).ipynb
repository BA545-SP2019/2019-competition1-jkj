```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
%matplotlib inline
plt.style.use('ggplot')
```

```python
# Uploading imputed data to df from CSV WorkingDataFrame
df = pd.read_csv('OutlierFix4', index_col=0)
df.head()
```

```python
df.mean()
```

```python
cols = ['C1', 'C2', 'C4', 'C7', "C3'", "C5'","C6'", "T4'", "T3'", "T5'", "S1'", "S2'", "S3'"]
from sklearn.preprocessing import MinMaxScaler
minmax = preprocessing.MinMaxScaler()
norm_df = (minmax.fit_transform(df))
norm_df = pd.DataFrame(norm_df, columns=cols)
```

```python
norm_df.head()
```

```python
# Observing the mean of the norm_df DataFrame 
norm_df.mean()
```

```python
# Exporting DataFrame as CSV 
norm_df.to_csv('MMNormDF4', ',')
```
